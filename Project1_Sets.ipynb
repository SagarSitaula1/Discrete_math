# Power Set

```
def DPpowerset(s):
  powerset = [[]]
  for i in s:
    for sett in powerset:
      powerset = powerset + [sett + [i]]
  return powerset    
```

After, using foor loop to define a power I am going to compute a power set for the set A.

```
DPpowerset(forwards)
```
